```python
# ðŸ“˜ Colab & Conquer â€” Your First Data Notebook

# âœ… Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Enable inline plotting
%matplotlib inline

# âœ… Step 2: Load the dataset
url = "https://bit.ly/4nejNue"
df = pd.read_csv(url)

# âœ… Step 3: Preview the data
print("ðŸ“„ First 5 rows of the dataset:")
display(df.head())

# âœ… Step 4: Basic Statistics
print("ðŸ“Š Summary Statistics:")
display(df.describe())

print("\nðŸ“Œ Mean of each feature:")
display(df.mean(numeric_only=True))

print("\nðŸ“Œ Median of each feature:")
display(df.median(numeric_only=True))

print("\nðŸ“Œ Standard Deviation of each feature:")
display(df.std(numeric_only=True))

# âœ… Step 5: Histograms of all numeric features
df.hist(figsize=(12, 8), bins=20, edgecolor='black')
plt.suptitle("ðŸ“‰ Histograms of Dataset Features", fontsize=16)
plt.tight_layout()
plt.show()
```
